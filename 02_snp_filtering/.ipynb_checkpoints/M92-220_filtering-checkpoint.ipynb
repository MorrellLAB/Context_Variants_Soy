## Plotting variant density

```python
def plot_windowed_variant_density(pos, window_size, title=None):
    
    # setup windows 
    bins = np.arange(0, pos.max(), window_size)
    
    # use window midpoints as x coordinate
    x = (bins[1:] + bins[:-1])/2
    
    # compute variant density in each window
    h, _ = np.histogram(pos, bins=bins)
    y = h / window_size
    
    # plot
    fig, ax = plt.subplots(figsize=(12, 3))
    sns.despine(ax=ax, offset=10)
    ax.plot(x, y)
    ax.set_xlabel('Chromosome position (bp)')
    ax.set_ylabel('Variant density (bp$^{-1}$)')
    if title:
        ax.set_title(title)
```

```python
title = 'Raw variant density over ' + chrom
pos = variants['POS'][:]
plot_windowed_variant_density(pos, window_size=100000, title=title)
```

## Ts/Tv Plots
Since you can't calculate Ts/Tv for triallelic sites or insertions/deletions, let's load a version of the VCF file that's only biallelic SNPs.
BCFtools was used to do the filtering (-m2 -M2 -v snps)

```python
new_file = '/panfs/roc/groups/9/morrellp/shared/Projects/Context_Of_Mutations/M92-220/Genotype_GVCFs/M92-220_complete_biallelic.vcf'
allel.vcf_to_hdf5(new_file,
                  '/panfs/roc/groups/9/morrellp/shared/Projects/Context_Of_Mutations/M92-220/Genotype_GVCFs/M92-220_' + chrom + '_biallelic.h5',
                  fields='*', 
                  region=chrom, 
                  overwrite=True)
```

```python
new_callset = h5py.File('/panfs/roc/groups/9/morrellp/shared/Projects/Context_Of_Mutations/M92-220/Genotype_GVCFs/M92-220_' + chrom + '_biallelic.h5', mode='r')
new_variants = allel.VariantChunkedTable(new_callset['variants'], 
                                     names=['POS', 'REF', 'ALT', 'DP', 'MQ', 'QD', 'AC'],
                                     index='POS')
#new_variants
```

```python
mutations = new_variants['REF'][:] + new_variants['ALT'][:,0]
#mutations
```

```python
def locate_transitions(x):
    x = np.asarray(x)
    return (x == 'AG') | (x == 'GA') | (x == 'CT') | (x == 'TC')
```

```python
is_ti = locate_transitions(mutations)
#is_ti
```

```python
def ti_tv(x):
    if len(x) == 0:
        return np.nan
    is_ti = locate_transitions(x)
    n_ti = np.count_nonzero(is_ti)
    n_tv = np.count_nonzero(~is_ti)
    if n_tv > 0:
        return n_ti / n_tv
    else:
        return np.nan
```

```python
ti_tv(mutations)
```

```python
def plot_ti_tv(f, downsample, bins):
    fig, ax = plt.subplots(figsize=(7, 5))
    sns.despine(ax=ax, offset=10)
    x = new_variants[f][:][::downsample]
    
    # plot a histogram
    ax.hist(x, bins=bins)
    ax.set_xlabel(f)
    ax.set_ylabel('No. variants')

    # plot Ti/Tv
    ax = ax.twinx()
    sns.despine(ax=ax, bottom=True, left=True, right=False, offset=10)
    values = mutations[::downsample]
    with np.errstate(over='ignore'):
        # binned_statistic generates an annoying overflow warning which we can ignore
        y1, _, _ = scipy.stats.binned_statistic(x, values, statistic=ti_tv, bins=bins)
    bx = (bins[1:] + bins[:-1]) / 2
    ax.plot(bx, y1, color='k')
    ax.set_ylabel('Ts/Tv')
    ax.set_ylim(0.5, 3)

    ax.set_title('Variant %s and Ts/Tv' % f)
```

```python
plot_ti_tv('QD', downsample=5, bins=np.arange(0, 40, 1))
```

```python
plot_ti_tv('MQ', downsample=5, bins=np.arange(0, 60, 1))
```

```python
plot_ti_tv('DP', downsample=5, bins=np.linspace(0, 800, 50))
```

```python
def plot_joint_ti_tv(f1, f2, downsample, gridsize=20, mincnt=20, vmin=0.5, vmax=2.5, extent=None):
    fig, ax = plt.subplots()
    sns.despine(ax=ax, offset=10)
    x = new_variants[f1][:][::downsample]
    y = new_variants[f2][:][::downsample]
    C = mutations[::downsample]
    im = ax.hexbin(x, y, C=C, reduce_C_function=ti_tv, mincnt=mincnt, extent=extent,
                   gridsize=gridsize, cmap='jet', vmin=vmin, vmax=vmax)
    fig.colorbar(im)
    ax.set_xlabel(f1)
    ax.set_ylabel(f2)
    ax.set_title('Variant %s versus %s and Ts/Tv' % (f1, f2))
```

```python
plot_joint_ti_tv('QD', 'MQ', downsample=1, mincnt=30)
```

```python
plot_joint_ti_tv('QD', 'DP', downsample=1, mincnt=30, extent=(0, 40, 0, 1000))
```

```python
plot_joint_ti_tv('MQ', 'DP', downsample=1, mincnt=30, extent=(0, 60, 0, 1000))
```

## Site Frequency Spectrum

```python
calldata = new_callset['calldata/GT']
genotypes = allel.GenotypeChunkedArray(calldata)
#genotypes
```

```python
ac = genotypes.count_alleles(max_allele=3)
#ac[0:50]
```

```python
is_biallelic = ac.is_biallelic_01()[:]
#is_biallelic
```

```python
len(is_biallelic)
```

```python
ac_bi = ac.compress(is_biallelic, axis=0)[:, :2]
#ac_bi
```

```python
print(str((len(is_biallelic) - len(ac_bi))/len(is_biallelic)*100) + 
    '% of variants are called as the alternative allele 100% of the time')
```

```python
ac_bi_alt = ac_bi[:, 1]
ac_bi_alt
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.despine(ax=ax, offset=10)
sfs1 = allel.sfs(ac_bi_alt)
allel.plot_sfs(sfs1, ax=ax, n=8)
#ax.set_xlim(0, 1)
#ax.set_ylim(0, 45000)
ax.set_title('Site Frequency Spectrum')
# workaround bug in scikit-allel re axis naming
ax.set_xlabel('Alternate allele frequency');
```

It's "unfolded" in the sense that we care about the frequency of the alternate allele.

If the alternate allele is at a high frequency then it's supported by multiple M92-220 samples, which means it's much more likely to be a real variant.

In the unfiltered dataset, there are a lot of singleton heterozygotes (freq = 0.1).

## Variant Filtering

```python
filter_expression = '(QD > 20) & (MQ > 35) & (DP > 5) & (DP < 500)'
variant_selection = new_variants.eval(filter_expression)[:]
```

Number of remaining variants:

```python
np.count_nonzero(variant_selection)
```

Percent of variants remaining:

```python
print(str(np.count_nonzero(variant_selection)/len(new_variants)*100) + ' %')
```

```python
variants_pass = new_variants.compress(variant_selection)
variants_pass
```

```python
genotypes_pass = genotypes.compress(variant_selection)
genotypes_pass
```

```python
mutations_pass = variants_pass['REF'][:] + variants_pass['ALT'][:,0]
ti_tv(mutations_pass)
```

```python
ac_pass = genotypes_pass.count_alleles(max_allele=3)
is_biallelic_pass = ac_pass.is_biallelic_01()[:]
ac_bi_pass = ac_pass.compress(is_biallelic_pass, axis=0)[:, :2]
ac_bi_alt_pass = ac_bi_pass[:, 1]
print(str((len(is_biallelic_pass) - len(ac_bi_pass))/len(is_biallelic_pass)*100) + 
    '% of variants are called as the alternative allele 100% of the time')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.despine(ax=ax, offset=10)
sfs1 = allel.sfs(ac_bi_alt_pass)
allel.plot_sfs(sfs1, ax=ax, n=10)
ax.set_xlim(0, 1)
#ax.set_ylim(0, 45000)
ax.set_title('Site Frequency Spectrum')
# workaround bug in scikit-allel re axis naming
ax.set_xlabel('Alternate allele frequency');
```

Scikit-allel won't accept the variants that are 100% the alternative allele, so keep in mind that this SFS only applies to 7% of the total variants.
